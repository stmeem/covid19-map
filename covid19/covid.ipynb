# COVID-19 World Map

```python
import folium
import numpy
import pandas
from IPython.core.display import display, HTML

data = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv')

latitude =list(data["Lat"].replace(numpy.nan,0))
longitude =list(data ["Long_"].replace(numpy.nan,0))
country = list(data["Country_Region"])
cases = list(data["Confirmed"])
recover = list(data["Recovered"])
death = list(data["Deaths"])
update=list(data["Last_Update"])
lastupdate=update[0][0:10]
display(HTML("<p> Last Updated: "+ str(lastupdate)+"</p>"))
```

```python

map = folium.Map(location = [30,112],zoom_start=2, tiles="Stamen Terrain")

feature_group = folium.FeatureGroup(name="My Map")

def color_icon(cases):
    if cases < 5000:
        return 'green'
    
    elif 5000 <= cases < 50000:
        return 'blue'
    
    else:
        return 'red'
    

for lat, lon, c,d,recovered, t_case in zip(latitude, longitude, country, death, recover, cases):
    feature_group.add_child(folium.Marker(
        location =[lat, lon], popup =str(c)+" Cases: "+str(t_case)+" Recovered: "+str(recovered)+" Death: "+ str(d), icon =folium.Icon(color=color_icon(t_case))))
    
map.add_child(feature_group)

map
```
